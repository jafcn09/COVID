# **REGRESION LOGISTICA**

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
```

```python
X_rl = df_proyecto[['EDAD_x', 'SEXO_x_FEMENINO', 'SEXO_x_MASCULINO','POSITIVO']]
y_rl = df_proyecto['Fallecio']
```

```python
X_train_rl,X_test_rl,y_train_rl,y_test_rl = train_test_split(X_rl,y_rl,test_size=0.3,random_state=42)
```

```python
scaler = MinMaxScaler()
X_train_rl[['EDAD_x']] = scaler.fit_transform(X_train_rl[['EDAD_x']])
X_test_rl[['EDAD_x']] = scaler.transform(X_test_rl[['EDAD_x']])
```

```python
log_regression = LogisticRegression(class_weight= {0:1, 1:10})
log_regression.fit(X_train_rl,y_train_rl)
```

```python
y_pred_rl = log_regression.predict(X_test_rl)
y_pred_rl
```

```python
cnf_matrix = metrics.confusion_matrix(y_test_rl, y_pred_rl)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test_rl, y_pred_rl))
```

```python
print(classification_report(y_test_rl, y_pred_rl, target_names=["Vivo", "Muerto"]))
```

```python
pred = log_regression.predict([[scaler.transform([[88]]), 0, 1, 1], [scaler.transform([[88]]), 0, 1, 0], [scaler.transform([[44]]), 0, 1, 0], [scaler.transform([[77]]), 0, 1, 1]])
pred
```

# **CLUSTERING**

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
df_proyecto.info()
```

```python
df_proyecto.shape
```

```python
print(df_proyecto.groupby('POSITIVO').size())
```

```python
df_proyectoR = df_proyecto[['DEPARTAMENTO_MUESTRA', 'POSITIVO', 'SEXO_x_FEMENINO', 'SEXO_x_MASCULINO']]
```

```python
departamentos = df_proyectoR.DEPARTAMENTO_MUESTRA.tolist()
```

```python
df_proyectoR = df_proyectoR.drop(['DEPARTAMENTO_MUESTRA'], axis = 1)
```

```python
xDep = df_proyectoR.values
scalerDep = preprocessing.MinMaxScaler()
xDep_scale = scalerDep.fit_transform(xDep)
xDep_norm = pd.DataFrame(xDep_scale)
```

```python
xDep_norm.head()
```

```python
pca = PCA(n_components = 2)
reduce_dep = pd.DataFrame(pca.fit_transform(xDep_norm))
```

```python
kmeans = KMeans(n_clusters= 2)
kmeans = kmeans.fit(reduce_dep)
labels = kmeans.predict(reduce_dep)
centroid = kmeans.cluster_centers_
clusters = kmeans.labels_.tolist()
```

```python
reduce_dep['cluster'] = clusters
reduce_dep['DEPARTAMENTOS'] = departamentos
reduce_dep.columns = ['FEMENINO', 'MASCULINO', 'POSITIVO', 'DEPARTAMENTOS']
reduce_dep.head()
```

```python
print(reduce_dep.groupby('POSITIVO').size())
```
